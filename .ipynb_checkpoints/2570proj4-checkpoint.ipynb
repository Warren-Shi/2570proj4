### Q1

```python
'''
Returns the theoretical price and delta of a digital option
'''
def digit_theo(K, St, r, sig, T, t):
    d1 = (np.log(St/K) + (r + sig**2/2)*(T-t))/(sig*np.sqrt(T-t))
    d2 = d1 - sig*np.sqrt(T-t)
    ## price
    c_digital = np.exp(-r*(T-t))*st.norm.cdf(d2)
    ## delta
    delta_digital_theo = np.exp(-r*(T-t))*st.norm.pdf(d2)/(sig*St*np.sqrt(T-t))
    return (c_digital, delta_digital_theo)

'''
forward difference method for estimating delta of digital option
'''
def digit_fwd_diff(K, St, r, sig, T, t, h, n):
    ## get n normal random variables as paths
    W = st.norm.rvs(size = n)
    ## calculate difference in payoff
    diff_payoff = (((St+h)*np.exp((r-1/2*sig**2)*(T-t)+sig*np.sqrt(T-t)*W) > K).astype(int) - 
                   ((St)*np.exp((r-1/2*sig**2)*(T-t)+sig*np.sqrt(T-t)*W) > K).astype(int)) 
    ## calculate difference in price
    diff_val = np.exp(-r*(T-t))*diff_payoff/(2*h)
    ## return the expected value (mean) as estimation
    return np.mean(diff_val)

def mse(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)
```

```python
## Provided parameters
K = 110
St = 100
r = 0.02
sig = 0.25
T = 0.25
t = 0
n = 5000
```

```python
## get true delta
y_true = digit_theo(K, St, r, sig, T, t)[1]
## h's we will be plotting over
h = np.linspace(0.01, 50, 1000)
## get mean square error for each h
fwd_errors = [mse(y_true, digit_fwd_diff(K, St, r, sig, T, t, hi, n)) for hi in h]
```

```python
## plot errors vs. h
plt.plot(h, fwd_errors)
plt.title('Mean Square Error of Forward Difference Method Against Step Size h')
plt.ylabel('MSE')
plt.xlabel('Step Size');
```

By eyeball estimation, the optimal h is about 13.

### Q2

```python
'''
simulate a GBM path for the Asian Options with a given RNG up to m weeks, repeat n times
'''
def sim_path(K, S0, r, sig, m, n, W):
    dt = 1/52
    ## simlate weekly returns
    weekly_returns = np.exp((r - 1/2*sig**2)*dt + sig*np.sqrt(dt)*W)
    ## simulate path
    Sti = S0 * np.cumprod(weekly_returns, axis = 1)
    return Sti

'''
Calculate the vegas of an Asian Option using pathwise derivative
'''
def asian_pathwise(K, S0, r, sig, m, n, paths):
    T = m/52
    ti = np.array([i/52 for i in range(1, m+1)])
    ti = np.tile(ti, (n, 1))
    indic =  (np.mean(paths, axis = 1) > K).astype(int)
    dSti = (np.log(paths/S0) - (r + 1/2*sig**2)*ti)*paths/sig
    return np.exp(-r*T)*np.sum(dSti, axis = 1)*indic/m

'''
Calculate the vegas of an Asian Option using likelihood ratio
'''
def asian_lr(K, S0, r, sig, m, n, paths, Z):
    T = m/52
    dt = 1/52
    indic =  (np.mean(paths, axis = 1) > K).astype(int)
    payoff = np.exp(-r*T)*(np.mean(paths, axis = 1) - K)*indic
    return payoff*np.sum((Z**2-1)/sig - Z*np.sqrt(dt), axis = 1)
```

```python
## Provided parameters
S0 = 100
K = 100
sig = 0.3
r = 0.05

n = 100000
```

```python
## simulate 
## malloc
var_pw = np.zeros(256)
mean_pw = np.zeros(256)
var_lr = np.zeros(256)
mean_lr = np.zeros(256)

## go through m's
for m in np.arange(1, 257):
    ## generate n normal paths of length m
    norms = np.random.normal(size = (n, m))
    ## generate n GBMs with length m
    paths = sim_path(K, S0, r, sig, m, n, norms)
    ## Pathwise derivative
    pathwise_sim = asian_pathwise(K, S0, r, sig, m, n, paths)
    mean_pw[m-1] = np.mean(pathwise_sim)
    var_pw[m-1] = np.log(np.var(pathwise_sim))/np.log(10)
    ## Likelihood ratio
    lr_sim = asian_lr(K, S0, r, sig, m, n, paths, norms)
    mean_lr[m-1] = np.mean(lr_sim)
    var_lr[m-1] = np.log(np.var(lr_sim))/np.log(10)
```

```python
plt.plot(mean_pw, label = 'pathwise')
plt.plot(mean_lr, label = 'likelihood')
plt.title('Estimated Vega')
plt.legend()
plt.ylabel('Estemation')
plt.xlabel('Number of Weeks');
```

```python
plt.plot(var_pw, label = 'pathwise')
plt.plot(var_lr, label = 'likelihood')
plt.title('Variance of Vega Estimation')
plt.legend()
plt.ylabel('10^n')
plt.xlabel('Number of Weeks');
```

### Q3

```python
xy_csv = pd.read_csv('data/XY.csv')[['X', 'Y']]
xy = xy_csv.values
rho_mle = np.corrcoef(xy.T)[0,1]
```

```python
def L(data, rho):
    mu = np.array([0,0])
    Sig = np.array([[1, rho], [rho, 1]])
    return np.prod(st.multivariate_normal(mu, Sig).pdf(data))

def pi(p):
    return (1-p**2)**(-3/2)
```

```python
rho_est = [rho_mle]
for i in range(50000):
    rho_new = np.random.normal(rho_mle, 0.2)
    if rho_new > 1 or rho_new < -1:
        continue
    rho_curr = rho_est[-1]
    L_new = L(xy, rho_new)
    L_curr = L(xy, rho_curr)
    r = L_new/L_curr*pi(rho_new)/pi(rho_curr)
    if r > 1:
        r = 1
    U = np.random.uniform()
    if U<=r:
        rho_est += [rho_new]
    else:
        rho_est += [rho_curr]
```

```python
plt.plot(rho_est)
plt.title('Metropolis-Hasting Sample Path');
```

***Note on Convergence:***  
We can see that the path is pretty consistant after 50000 iterations. It is likely that our simulation has already, or at least has almost converged. Higher number of iteration is also computationally expensive, so we stop at 50000 iterations

```python
sns.distplot(rho_est)
plt.title('Distribution of H-S Sample Path');
```

```python
hs_mode = st.mode(rho_est)[0][0]
print('The MAP estimate is: ' + str(hs_mode))
print('The 95% credible interval is: ' + str((np.quantile(rho_est, 0.025), np.quantile(rho_est, 0.975))))
```

### Q4

```python
from sklearn.linear_model import LinearRegression
```

```python
inflation = pd.read_csv('data/inflation.csv')['Inflation']
Y = np.roll(inflation, -2)[:250]
X = np.vstack([np.ones(252), np.roll(inflation, -1), inflation])[:, :250].T
```

```python
'''
gibbs sampling
'''
def gibbs(X, Y, reps):
    ## first pass
    ## estimate first sigma squared 
    reg = LinearRegression().fit(X, Y)
    res = reg.predict(X) - Y
    sig_sq0 = np.var(res)
    
    gibb_paths = np.zeros((reps, 4))
    sig_sq = sig_sq0
    mu = np.linalg.inv(X.T.dot(X)/sig_sq).dot(X.T.dot(Y)/sig_sq)
    Sigma = np.linalg.inv(X.T.dot(X)/sig_sq)
    beta_curr = st.multivariate_normal.rvs(mean = mu, cov = Sigma)
    gibb_paths[0] = np.hstack([beta_curr, sig_sq])
    
    for i in range(1, reps):
        ## inverse gamma
        sim_invg = np.random.normal(size = 250)
        D1=0.1*0.5+res.dot(res)
        sig_sq = D1/sim_invg.dot(sim_invg)
        ## normal
        mu = np.linalg.inv(X.T.dot(X)/sig_sq).dot(X.T.dot(Y)/sig_sq)
        Sigma = np.linalg.inv(X.T.dot(X)/sig_sq)
        beta_curr = st.multivariate_normal.rvs(mean = mu, cov = Sigma)
        gibb_paths[i] = np.hstack([beta_curr, sig_sq])
        
    return gibb_paths


'''
forecast future values with a given gibbs sampling path
'''
def forecast(coef):
    y251 = float(X[-1,:].dot(coef[:3]) + np.random.normal(size=1,loc=0,scale=coef[3]))
    y252 = float(np.array([1,y251,Y[-1]]).dot(coef[:3]) + np.random.normal(size=1,loc=0,scale=coef[3]))
    y253 = float(np.array([1,y252,y251]).dot(coef[:3]) + np.random.normal(size=1,loc=0,scale=coef[3]))
    y254 = float(np.array([1,y253,y252]).dot(coef[:3]) + np.random.normal(size=1,loc=0,scale=coef[3]))
    return np.array([y251, y252, y253, y254])
```

```python
## simulation
gibbs_result = gibbs(X, Y, reps = 6000)[4000:]
gibbs_mean = np.mean(gibbs_result, axis = 0)
```

```python
sns.distplot(gibbs_result[:, 0])
plt.title('Distribution Plot of Estimated alpha');
```

#### C). 

```python
## Forecast
fresults=np.zeros((2000,4))
for i in range(2000):
    fresults[i,:]=forecast(gibbs_result[i,:])
preds = np.median(fresults,axis=0)
lower = np.quantile(fresults,0.025,axis=0)
upper = np.quantile(fresults,0.975,axis=1)

k = 1
for i in zip(lower, upper):
    print('The MAP estimate for Q' + str(k) + ' is: ' + str(preds[k-1]))
    print('The 95% credible intervals for Q' + str(k) + ' is: ' + str(i))
    k += 1
```
